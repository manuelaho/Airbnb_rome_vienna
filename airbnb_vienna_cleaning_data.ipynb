# Null values

```python
df_vienna.isnull().sum()
```

```python
# Delete features "neighbourhood_group" and "license"

df_vienna.drop("neighbourhood_group", axis=1, inplace=True)

df_vienna.drop("license", axis=1, inplace=True)
```

```python
# Delete feature "last_review"

df_vienna.drop("last_review", axis=1, inplace=True)
```

```python
# Delete features "id" and "host_id" --> not important for analysies

df_vienna.drop("id", axis=1, inplace=True)

df_vienna.drop("host_id", axis=1, inplace=True)
```

```python
# Delete 1 sample of null values from column "host_name":

df_vienna.drop(df_vienna[df_vienna["host_name"].isnull()].index, inplace=True)
```

```python
#Column "reviews_per_month" --> fill null values with mean:

df_vienna["reviews_per_month"].fillna(df_vienna["reviews_per_month"].mean(),inplace=True)
```

# Checking outliers

```python
# Checking minimum_nights > 30:

df_vienna[df_vienna["minimum_nights"]  > 30]

#Delete  minimum_nights > 30:
    
df_vienna.drop(df_vienna[df_vienna["minimum_nights"] > 30].index, inplace=True)
```
